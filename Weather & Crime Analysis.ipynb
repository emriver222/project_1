```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
import random
from datetime import datetime
```

### READ IN FILES

```python
weather_path = "Output/Weather_AvgTemp_Monthly"
weather_data = pd.read_csv(weather_path)
weather_mo_df = pd.DataFrame(weather_data)
#weather_mo_df.head()
```

```python
crime_path = "Output/CrimeData_Monthly.csv"
crime_data = pd.read_csv(crime_path)
crime_mo_df = pd.DataFrame(crime_data)
```

```python
weather_mo_df = weather_mo_df.rename(columns= {"ReportedYearMonth":"Month"})
#weather_mo_df.head()
```

```python
#Combined Crime & Weather Monthly DataFrame
combined_monthly = pd.merge(crime_mo_df, weather_mo_df, on="Month")
#combined_monthly
```

```python
#Daily Crime
daily_crime = "Output/CrimeData_Daily.csv"
daily_crime_df = pd.read_csv(daily_crime, encoding="ISO-8859-1")
```

```python
#Crime
jan2019_df = daily_crime_df.loc[daily_crime_df["Day"] >= "2019-01-01"]
jan2019_df = jan2019_df.loc[jan2019_df["Day"] < "2019-02-01"]
```

```python
#Crime
jul2019_df = daily_crime_df.loc[daily_crime_df["Day"] >= "2019-07-01"]
jul2019_df = jul2019_df.loc[jul2019_df["Day"] < "2019-08-01"]
```

```python
#Daily Weather
daily_weather = "Output/Weather_AvgTemp_Daily"
daily_weather = pd.read_csv(daily_weather)
daily_weather = daily_weather.rename(columns = {"ReportedYearDay":"Day"})
#daily_weather
```

```python
#Weather
weath_jan2019 = daily_weather.loc[daily_weather["Day"] >= "2019-01-01"]
weath_jan2019 = weath_jan2019.loc[weath_jan2019["Day"] < "2019-02-01"]
```

```python
#Weather
weath_jul2019 = daily_weather.loc[daily_weather["Day"] >= "2019-07-01"]
weath_jul2019 = weath_jul2019.loc[weath_jul2019["Day"] < "2019-08-01"]
```

```python
#Combined Crime & Weather Daily January 19' & July 19' DataFrames
combined_daily_JAN = pd.merge(jan2019_df, weath_jan2019, on="Day")
combined_daily_JUL = pd.merge(jul2019_df, weath_jul2019, on="Day")
combined_daily_JUL = combined_daily_JUL.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'])
combined_daily_JAN = combined_daily_JAN.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'])
```

### CREATE MONTHLY VISUALS

```python
mo_linechart = combined_monthly.set_index("Month")
#mo_linechart.head()
```

```python
fig, ax1 = plt.subplots()

x_axis = np.arange(len(combined_monthly['Month']))

color = 'tab:blue'
ax1.set_xlabel('Month')
ax1.set_ylabel('Avg Temp', color=color)
ax1.plot(combined_monthly['Month'], combined_monthly['Avg_temp'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

fig.autofmt_xdate(rotation=90)

ax2 = ax1.twinx() 

color = 'tab:red'
ax2.set_ylabel('Crime Count', color=color) 
ax2.plot(combined_monthly['Month'], combined_monthly['CrimeCount'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

tick_locations = [value for value in x_axis]

plt.title("Monthly Average Temperature vs. Crime Count")
plt.show()
```

```python
temp = combined_monthly['Avg_temp']
crime = combined_monthly['CrimeCount']

x_values = temp
y_values = crime

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values)

plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(10,10),fontsize=15,color="red")

plt.xlabel ('Temperature')
plt.ylabel ('Crime')
plt.title ('Temperature vs. Crime Count ')
plt.show()
print(line_eq)
```
